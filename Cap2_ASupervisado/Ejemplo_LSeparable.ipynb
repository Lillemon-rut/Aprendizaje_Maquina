```python
import numpy as np
import matplotlib.pyplot as plt

Grupo1=set([(0,1.8),(2,0.6)])
Grupo2=set([(-1.2,1.4),(0.4,-1)])


# Graficamos los puntos
Puntos_P = np.array([(0,1.8),(2,0.6)])
Puntos_N = np.array([(-1.2,1.4),(0.4,-1)])
for x in Puntos_P:
  plt.scatter(x[0],x[1],c='r')
for x in Puntos_N:
  plt.scatter(x[0],x[1],c='b')


def recta(x,w):
  return -x*w[0]/w[1]

dominio = np.linspace(-2,3,100)




def Perceptron(Datos1,Datos2):
  w = np.array([1,1])
  Conjunto1= set()
  Conjunto2= set()
  while Conjunto1 != Datos1 or Conjunto2 != Datos2:
    for x in Datos1:
      if w@x <= 0:
        w=w+x
      else:
        Conjunto1.add(x)
    for x in Datos2:
      if w@x > 0:
        w=w-x
      else:
        Conjunto2.add(x)
  return w

plt.plot(dominio, recta(dominio,Perceptron(Grupo1,Grupo2)),c='g')
plt.show()
```
